# The .swapaxes() Method

```python
p = pd.read_hdf('CODES/companies.h5', 'key')
p
```

```python
p.swapaxes('items', 'minor')
p3 = p.swapaxes(0, 2)
```

```python
p3.axes
```

```python
p3['MSFT'].head()
```

```python
p3.minor_xs('Close').head()
```
