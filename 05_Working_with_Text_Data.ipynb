# The .str.replace() Method

```python
chicago = pd.read_csv('Codes/chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.tail(3)
```

```python
chicago['Department'] = chicago['Department'].str.replace('MGMNT', 'MANAGEMENT')
chicago['Department'] = chicago['Department'].str.replace('SAN', 'SANITATION')
chicago.head(10)
```

```python
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$', '').astype('float')
chicago.head(5)
```

```python
chicago['Employee Annual Salary'].mean()
```

# Filtering with String Methods

```python
chicago = pd.read_csv('Codes/chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.tail(3)
```

# The expand and n Parameters of the str.split() Method

```python
chicago = pd.read_csv('Codes/chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.head(3)
```

```python
chicago[['First Name', 'Last Name']] = chicago['Name'].str.split(',', expand=True)
chicago.head(3)
```

```python
chicago[['First Title Word', 'Remaining']] = chicago['Position Title'].str.split(' ', expand=True, n=1)
chicago.head(3)
```
