```python
URL = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv'
baby_names = pd.read_csv(URL)
baby_names.to_hdf('CODES/baby_names.h5', 'baby_names')
```

# Export CSV File with the .to_csv() Method

```python
baby_names = pd.read_hdf('CODES/baby_names.h5', 'baby_names')
baby_names.head(3)
```

```python
baby_names.to_csv('CODES/NYC Baby Names.csv', index=False, 
                  columns=['Year of Birth', 'Child\'s First Name', 'Count'], 
                  encoding='utf-8')
```

# Export Excel File

```python
baby_names = pd.read_hdf('CODES/baby_names.h5', 'baby_names')
baby_names.head(3)
```

```python
girls = baby_names[baby_names['Gender'] == 'FEMALE']
boys = baby_names[baby_names['Gender'] == 'MALE']
```

```python
excel_file = pd.ExcelWriter('CODES/Baby Names.xlsx')
```

```python
girls.to_excel(excel_file, sheet_name='Girls', index=False)
boys.to_excel(excel_file, sheet_name='Boys', index=False, columns=['Gender', 'Child\'s First Name', 'Count'])
```

```python
excel_file.save()
```
