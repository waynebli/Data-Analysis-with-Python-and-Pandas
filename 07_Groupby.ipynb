# The .agg() Method

```python
fortune = pd.read_csv('Codes/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby(by='Sector')
fortune.head(3)
```

```python
sectors.agg({'Revenue' : ['sum', 'mean'], 'Profits' : 'sum', 'Employees' : 'mean'})
```

```python
sectors.agg(['size', 'sum', 'mean'])
```

# Iterating through Groups

```python
fortune = pd.read_csv('Codes/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby(by='Sector')
fortune.head(3)
```

```python
sectors['Profits'].max()
```

```python
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_in_company = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue_in_company)
```
