# The left_on and right_on Parameters

```python
week1 = pd.read_csv('Codes/Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Codes/Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Codes/Restaurant - Customers.csv')
foods = pd.read_csv('Codes/Restaurant - Foods.csv')
```

```python
week2.head(3)
```

```python
customers.head(3)
```

```python
week2 = week2.merge(customers, how='left', left_on='Customer ID', right_on='ID', sort=True).drop('ID', axis=1)
week2.head(3)
```

# Merging by Indexes with the left_index and right_index Parameters

```python
week1 = pd.read_csv('Codes/Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Codes/Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Codes/Restaurant - Customers.csv', index_col='ID')
foods = pd.read_csv('Codes/Restaurant - Foods.csv', index_col='Food ID')
```

```python
foods.head(3)
```

```python
customers.head(3)
```

```python
sales = week1.merge(customers, how='left', left_on='Customer ID', right_index=True)
sales.head(3)
```

```python
sales_items = sales.merge(foods, how='left', left_on="Food ID", right_index=True)
sales_items.head(3)
```

```python
week1.merge(week2, how='left', left_index=True, right_index=True, suffixes=[' - Week 1', ' - Week 2'])
```
